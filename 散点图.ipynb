```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,20)
plt.style.use('seaborn-whitegrid')
```

```python
plt.plot(x,np.sin(x),'o',color='black')
```

```python
fig = plt.figure()
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','+','v','^','<','>','s','d']:
    plt.plot(rng.rand(5),rng.rand(5),marker,
             label="marker='{}'".format(marker))
    plt.legend(numpoints = 1)
    plt.xlim(0,1.8)
fig.savefig('散点图.png')
```

```python
plt.plot(x,np.cos(x),'-ok')
```

```python
plt.plot(x,np.sin(x),'-p',color='gray',
         markersize=10,linewidth=2,
         markerfacecolor='white',markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2,1.2)
```

```python
plt.scatter(x,np.sin(x),marker='o')
```

```python
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar()
plt.savefig('分布图.png')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
feature = iris.data.T

plt.scatter(feature[0],feature[1],alpha=0.2,s=100*feature[3],
            c=iris.target,cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.savefig('鸢尾花.png')
```
