```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
plt.show()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```python
ax = plt.axes()
ax.plot(x,np.cos(x))
```

```python
ax = plt.axes()
ax.plot(x,np.cos(x))
ax.plot(x,np.sin(x + np.pi))
```
